```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load CSV File
valle_nevado_ml_df = pd.read_csv('./Valle_Nevado_SnowFall.csv')
valle_nevado_ml_df
```

```python
# Drop the last row, as 2022 isn't finished yet
valle_nevado_ml_df.drop(valle_nevado_ml_df.tail(1).index, inplace=True)
valle_nevado_ml_df
```

```python
# Visually Inspect the relationship between Snowfall and Years
plt.scatter(valle_nevado_ml_df.year, valle_nevado_ml_df.total_snowfall)
plt.xlabel('Year')
plt.ylabel('Snowfall(in inches)')
plt.show()
```

```python
# Format data to meet requirements of the Scikit-learn library
X = valle_nevado_ml_df.year.values.reshape(-1, 1)
```

```python
# Assign the y value
y = valle_nevado_ml_df.total_snowfall
```

```python
# Create an instance of the linear regression model
model = LinearRegression()
```

```python
# Fit the model
model.fit(X,y)
```

```python
# Generate predictions
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
# Plot the predictions as a red line against the data points
plt.scatter(X,y)
plt.plot(X, y_pred, color='red')
plt.title("Predicting Snowfall in Valle Nevado")
plt.xlabel('Year')
plt.ylabel('Snowfall(in inches)')
plt.show()
```

```python
#Examine slope and y-intercept
print(model.coef_)
print(model.intercept_)
```
