```python
# Import Dependencies
import pandas as pd
```

```python
# Read CSV as df
tiffindell_df = pd.read_csv('Tiffindell_Ski_Resort.csv')
```

```python
tiffindell_df
```

```python
tiffindell_df = tiffindell_df.drop(columns=['timezone', 'lat', 'lon', 'visibility', 'dew_point', 'feels_like', 'pressure', 'sea_level', 'grnd_level', 'wind_speed', 'wind_deg', 'wind_gust', 'rain_1h', 'rain_3h', 'snow_3h', 'clouds_all', 'weather_id', 'weather_icon'])
tiffindell_df
```

```python
# Fill Nan with 0
tiffindell_df = tiffindell_df.fillna(0)
tiffindell_df
```

```python
# Check info
tiffindell_df.info()
```

```python
# Convert dt to usable Datetime for index
datetime_info = pd.to_datetime(tiffindell_df['dt'], unit='s')
```

```python
# Set the index as datetime
tiffindell_datetime_df = tiffindell_df.set_index(datetime_info)
tiffindell_datetime_df
```

```python
# Reduce rows to show per day instead of per hour while getting rid of unnessessary columns
resampled_df = tiffindell_datetime_df.resample('D').agg({'temp': 'mean', 'temp_min' : 'mean', 'temp_max' : 'mean', 'humidity' : 'mean', 'snow_1h' : 'sum'})
resampled_df
```

```python
# Rename snow_1 to snowfall
tiffindell_cleaned_df = resampled_df.rename(columns={"snow_1h" : "snowfall"})
tiffindell_cleaned_df
```

```python
# Round the dataframe to 2 decimal places
tiffindell_cleaned_df = tiffindell_cleaned_df[['temp', 'temp_min', 'temp_max', 'humidity', 'snowfall']].round(2)
tiffindell_cleaned_df
```

```python
# Export to CSV
tiffindell_cleaned_df.to_csv('./cleaned_tiffindell.csv')
```
