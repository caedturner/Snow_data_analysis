```python
# Import dependencies
import pandas as pd
```

```python
# Read CSV as df
zermatt_df = pd.read_csv('./Zermatt_Ski_Resort.csv')
```

```python
zermatt_df
```

```python
# Drop Irrelevant columns
zermatt_df = zermatt_df.drop(columns=['timezone', 'lat', 'lon', 'visibility', 'dew_point', 'feels_like', 'pressure', 'sea_level', 'grnd_level', 'wind_speed', 'wind_deg', 'wind_gust', 'rain_1h', 'rain_3h', 'snow_3h', 'clouds_all', 'weather_id', 'weather_icon'])
zermatt_df
```

```python
# Fill Nan with 0
zermatt_df = zermatt_df.fillna(0)
zermatt_df
```

```python
#Check info
zermatt_df.info()
```

```python
# Convert dt to usable Datetime for index
datetime_info = pd.to_datetime(zermatt_df['dt'], unit='s')
```

```python
# Set the index as datetime
zermatt_datetime_df = zermatt_df.set_index(datetime_info)
zermatt_datetime_df
```

```python
# Reduce rows to show per day instead of per hour while getting rid of unnessessary columns
resampled_df = zermatt_datetime_df.resample('D').agg({'temp': 'mean', 'temp_min' : 'mean', 'temp_max' : 'mean', 'humidity' : 'mean', 'snow_1h' : 'sum'})
resampled_df
```

```python
# Rename snow_1 to snowfall
zermatt_cleaned_df = resampled_df.rename(columns={"snow_1h" : "snowfall"})
zermatt_cleaned_df
```

```python
# Round the dataframe to 2 decimal places
zermatt_cleaned_df = zermatt_cleaned_df[['temp', 'temp_min', 'temp_max', 'humidity', 'snowfall']].round(2)
zermatt_cleaned_df
```

```python
# Export to CSV
zermatt_cleaned_df.to_csv('./cleaned_zermatt.csv')
```
