```python
# Import dependancies 
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
# Locate data
Furano_Ski_Resort_to_load = "C:\\Users\\haley\\Desktop\\Snow_data_analysis\\Resources\\Furano_Ski_Resort.csv"
```

```python
#Read the Craigieburn_Valley data
Furano_Ski_Resort_df = pd.read_csv(Furano_Ski_Resort_to_load)
```

```python
Furano_Ski_Resort_df
```

```python
#Drop irrevelant columns
Furano_Ski_Resort_df = Furano_Ski_Resort_df.drop(columns=['timezone', 'lat', 'lon', 'dew_point', 'feels_like', 'pressure', 'sea_level', 'grnd_level', 'wind_speed', 'wind_deg', 'wind_gust', 'rain_1h', 'rain_3h', 'snow_3h', 'clouds_all', 'weather_id', 'weather_icon'])
```

```python
Furano_Ski_Resort_df = Furano_Ski_Resort_df.fillna(0)
```

```python
Furano_Ski_Resort_df
```

```python
# Convert dt to usable Datetime for index
datetime_info = pd.to_datetime(Furano_Ski_Resort_df['dt'],unit='s')
```

```python
# Set the index as datetime
Furano_Ski_Resort_datetime_index = Furano_Ski_Resort_df.set_index(datetime_info)
Furano_Ski_Resort_datetime_index
```

```python
# Resample data to reduce 24 hourly rows to 1 daily row - also remove columns not needed 
resampled_df = Furano_Ski_Resort_datetime_index.resample('D').agg({'temp': 'mean', 'temp_min' : 'mean', 'temp_max' : 'mean', 'humidity' : 'mean', 'snow_1h' : 'sum'})
```

```python
# Rename 'snow_1h' column to 'snowfall'
finished_df = resampled_df.rename(columns={"snow_1h" : "snowfall"})
```

```python
finished_df
```

```python
finished_df.info()
```

```python
# Round all columns to two decimal places
finished_df = finished_df[['temp', 'temp_min', 'temp_max', 'humidity', 'snowfall']].round(2)
finished_df
```

```python
# Export to CSV
finished_df.to_csv('./cleaned_Furano_Ski_Resort_df.csv')
```
