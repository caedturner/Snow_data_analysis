```python
# Import dependancies 
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
# Locate data
Craigieburn_Valley_to_load = "C:\\Users\\haley\\Desktop\\Snow_data_analysis\\Resources\\Craigieburn_Valley.csv"
```

```python
#Read the Craigieburn_Valley data
Craigieburn_Valley_df = pd.read_csv(Craigieburn_Valley_to_load)
```

```python
Craigieburn_Valley_df
```

```python
#Drop irrevelant columns
Craigieburn_Valley_df = Craigieburn_Valley_df.drop(columns=['timezone', 'lat', 'lon', 'dew_point', 'feels_like', 'pressure', 'sea_level', 'grnd_level', 'wind_speed', 'wind_deg', 'wind_gust', 'rain_1h', 'rain_3h', 'snow_3h', 'clouds_all', 'weather_id', 'weather_icon'])
```

```python
Craigieburn_Valley_df = Craigieburn_Valley_df.fillna(0)
```

```python
Craigieburn_Valley_df
```

```python
# Convert dt to usable Datetime for index
datetime_info = pd.to_datetime(Craigieburn_Valley_df['dt'],unit='s')
```

```python
# Set the index as datetime
Craigieburn_Valley_datetime_index = Craigieburn_Valley_df.set_index(datetime_info)
Craigieburn_Valley_datetime_index
```

```python
# Resample data to reduce 24 hourly rows to 1 daily row - also remove columns not needed 
resampled_df = Craigieburn_Valley_datetime_index.resample('D').agg({'temp': 'mean', 'temp_min' : 'mean', 'temp_max' : 'mean', 'humidity' : 'mean', 'snow_1h' : 'sum'})
```

```python
# Rename 'snow_1h' column to 'snowfall'
finished_df = resampled_df.rename(columns={"snow_1h" : "snowfall"})
```

```python
finished_df
```

```python
finished_df.info()
```

```python
# Round all columns to two decimal places
finished_df = finished_df[['temp', 'temp_min', 'temp_max', 'humidity', 'snowfall']].round(2)
finished_df
```

```python
# Export to CSV
finished_df.to_csv('./cleaned_Craigieburn_Valley_df.csv')
```
